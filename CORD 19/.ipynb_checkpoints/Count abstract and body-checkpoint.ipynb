```python
def body_text_sha(sha, lower = True):
    
    # The path where the files are saved
    path = "C:\\Users\\User\\OneDrive - NTNU\\NTNU\\Prosjekt oppgave NLP\\dataset\\CORD-19\\2020-10-18\\document_parses\\pdf_json\\"
    

    body_text = ""
        
    for s in sha:
        body_text_temp = ""
        try:
            with open(path+s+".json") as d:
                doc = json.load(d)
                n = len(doc["body_text"])
                for i in range(n):
                    if lower:
                        body_text_temp += " " + doc["body_text"][i]["text"].lower()
                    else:
                        body_text_temp += " " + doc["body_text"][i]["text"]
        except IOError:
            return ""
        if len(body_text_temp) > len(body_text):
            body_text = body_text_temp
    return body_text
    
```
